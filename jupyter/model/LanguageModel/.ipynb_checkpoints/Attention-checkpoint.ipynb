````python
def get_angles(pos, i, d_model):
    '''
    pos :     
    i :     
    d_model :        
    
    Positional Embedding  pos / 10000 ** (2i / d_model)   

    PE(pos, 2i) = sin(pos / 10000 ** (2i / d_model))     ((pos, 2i) sin))
    PE(pos, 2i + 1) = cos(pos / 10000 ** (2i / d_model)) ((pos, 2i+1 cos))
    '''
    angle_rates = 1 / np.power(10000, (2 * i // 2) / np.float32(d_model))
    # shape = (1, d_model)
    # pos shape :
    # angle_rates shape = (1, d_model)
    # return shape =  (pos, 1), (1, d_model) =  (pos, d_model)
    return np.matmul(pos, angle_rates)
````

````python
def positional_encoding(position, d_model):
    angle_rads = get_angles(np.arange(position)[:, np.newaxis],
                            np.arange(d_model)[np.newaxis, :],
                            d_model)  # shape : (position, d_model)
    #    sin  
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    #    cos  
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])

    pos_encoding = angle_rads[np.newaxis, ...]  # pos_encoding shape : (1, position, d_model)
    #  shape 1 , batch_size   

    return tf.cast(pos_encoding, dtype=tf.float32)
````

````python
class PositionalEncoding(tf.keras.layers.Layer):
    '''
          ,         
                 .
             .
    
    position :     
    i :     
    d_model :        
    
    Positional Embedding  position / 10000 ** (2i / d_model)   

    PE(position, 2i) = sin(position / 10000 ** (2i / d_model))     ((pos, 2i) sin))
    PE(position, 2i + 1) = cos(position / 10000 ** (2i / d_model)) ((pos, 2i+1 cos))
    
    shape = (1, d_model)
    angle_rates shape = (1, d_model)
    return shape =  (position, 1), (1, d_model) =  (position, d_model)
    '''
    def __init__(self, position, d_model):
        super(PositionalEncoding, self).__init__()
        self.pos_encoding = self.positional_encoding(position, d_model)
    def get_angles(self, position, i, d_model):

        angles = 1 / tf.pow(10000, (2 * (i // 2)) / tf.cast(d_model, tf.float32))
        return position * angles

    def positional_encoding(self, position, d_model):
        angle_rads = self.get_angles(
        position= tf.range(position, dtype=tf.float32)[:, tf.newaxis],
        i=tf.range(d_model, dtype=tf.float32)[tf.newaxis, :],
        d_model=d_model)# shape : (position, d_model)
        #    sin  
        sines = tf.math.sin(angle_rads[:, 0::2])
        #    cos  
        cosines = tf.math.cos(angle_rads[:, 1::2])
        
        angle_rads = np.zeros(angle_rads.shape)
        angle_rads[:, 0::2] = sines
        angle_rads[:, 1::2] = cosines
        
        pos_encoding = tf.constant(angle_rads)
        pos_encoding = angle_rads[tf.newaxis, ...]  # pos_encoding shape : (1, position, d_model)

        return tf.cast(pos_encoding, dtype=tf.float32)

    
    def call(self, inputs):
        return inputs + self.pos_encoding[:, :tf.shape(inputs)[1], :]
````

### Self-Attention
**Attention         .                 .**



1. Q, K, V  
    -     Q, K, V   .
    - Q, K, V   d_model        num_heads 
        -         .
        -    d_model*(d_model/num_heads)  .
        -  , d_model=512 num_heads=8   3     64   Q, K, V  .

2. Scaled dot-product Attention
    -  Q  K    ,        V         .    Q  .
        - Scaled dot-product Attention : $score(q,k) = \frac{q*k} {\sqrt{n}}$ (dot-product Attention    )       .
        -           V    .

3.    .
    -  Scaled dot-product Attention    .
    
    -              Q, K, V  .

    
        - Q K      Q K        .
        -     $\sqrt{d_k}$           .


        -        V   (Attention Value)  .


$$Attention(Q,K,V)= softmax(\frac{QK^T}{\sqrt{d_k}})V $$
    

````python
def scaled_dot_product_attention(q, k, v, mask=None):
    # query  : (batch_size, num_heads, query  , d_model/num_heads)
    # key  : (batch_size, num_heads, key  , d_model/num_heads)
    # value  : (batch_size, num_heads, value  , d_model/num_heads)
    # padding_mask : (batch_size, 1, 1, key  )
    
    # Q K . Attention Score .
    matmul_qk = tf.matmul(q, k, transpose_b = True)
    
    # scaling
    # d_k  .
    dk = tf.cast(tf.shape(k)[-1], tf.float32)
    scaled_attention_logits = matmul_qk / tf.math.sqrt(dk)
    
    # Masking
    #          .
    #           0 .
    if mask is not None:
        scaled_attention_logits += (mask * -1e9)

    #     k    .
    # attention weight shape : (batch_size, num_heads, query  , key  )
    attention_weights = tf.nn.softmax(scaled_attention_logits, axis=-1)

    output = tf.matmul(attention_weights, v)

    # output(attention_value) shape : (batch_size, seq_len, d_model/num_heads)
    #     (batch_size, seq_len, d_model/num_heads)   
    #    .

    # scaled_dot_product_attention      .
    return output, attention_weights
````

### Multi-head Attention



    -   num_heads=8,  8  ,        .
    -       .
    -        (concatenate). -> shape : (seq_len, d_model)
    -        $W^o$     -   .
    -          $W^o$  .
    -   -       (seq_len, d_model)  .
    -    -                  .
    -                   .
    



-    

        1. WQ, WK, WV  d_model  (Dense layer) .
        2.   (num_heads) (split).
        3.    .
        4.   (concatenatetion).
        5. WO    . 

````python
class MultiHeadAttention(tf.keras.layers.Layer):
    #          .
    #   ,      ,            .

    def __init__(self, d_model, num_heads, name="multi_head_attention"):
        super(MultiHeadAttention, self).__init__(name=name)
        self.num_heads = num_heads
        self.d_model = d_model

        assert d_model % self.num_heads == 0

        # d_model num_heads  .
        #   : 64
        self.depth = d_model // self.num_heads

        # WQ, WK, WV   
        self.query_dense = tf.keras.layers.Dense(units=d_model)
        self.key_dense = tf.keras.layers.Dense(units=d_model)
        self.value_dense = tf.keras.layers.Dense(units=d_model)

        # WO   
        self.dense = tf.keras.layers.Dense(units=d_model)

  # num_heads  q, k, v split 
    def split_heads(self, inputs, batch_size):
        inputs = tf.reshape(
            inputs, shape=(batch_size, -1, self.num_heads, self.depth))
        return tf.transpose(inputs, perm=[0, 2, 1, 3])

    def call(self, inputs):
        query, key, value, mask = inputs['query'], inputs['key'], inputs[
            'value'], inputs['mask']
        batch_size = tf.shape(query)[0]

        # 1. WQ, WK, WV   
        # q : (batch_size, query  , d_model)
        # k : (batch_size, key  , d_model)
        # v : (batch_size, value  , d_model)
        # ) (k, v)-(q)  query  key, value    .
        query = self.query_dense(query)
        key = self.key_dense(key)
        value = self.value_dense(value)

        # 2.  
        # q : (batch_size, num_heads, query  , d_model/num_heads)
        # k : (batch_size, num_heads, key  , d_model/num_heads)
        # v : (batch_size, num_heads, value  , d_model/num_heads)
        query = self.split_heads(query, batch_size)
        key = self.split_heads(key, batch_size)
        value = self.split_heads(value, batch_size)

        # 3.    .    .
        # (batch_size, num_heads, query  , d_model/num_heads)
        scaled_attention, _ = scaled_dot_product_attention(query, key, value, mask)
        # (batch_size, query  , num_heads, d_model/num_heads)
        scaled_attention = tf.transpose(scaled_attention, perm=[0, 2, 1, 3])

        # 4.  (concatenate)
        # (batch_size, query  , d_model)
        concat_attention = tf.reshape(scaled_attention,
                                      (batch_size, -1, self.d_model))

        # 5. WO   
        # (batch_size, query  , d_model)
        outputs = self.dense(concat_attention)

        return outputs
````

````python
class Pointwise_FeedForward_Network(tf.keras.layers.Layer):
    # Pointwise_FeedForward_Network    512  2048 ,  512  .
    def __init__(self, d_model, dff):
        super().__init__()
        self.d_model = d_model
        self.dff = dff

        self.middle = tf.keras.layers.Dense(dff, activation='relu')
        self.out = tf.keras.layers.Dense(d_model)

    def __call__(self, x):
        middle = self.middle(x) # middle shape : (batch_size, seq_len, dff)
        out = self.out(middle) # out shape : (batch_size, seq_len, d_model)
        return out
````

````python
class TransformerEmbedding(tf.keras.layers.Layer):
    def __init__(self, d_model, input_vocab_size, maximum_position_encoding, dropout_rate=0.1):
        super().__init__()

        self.d_model = d_model  #   d_model   
        self.embedding = tf.keras.layers.Embedding(input_vocab_size, d_model)
        # vocab_size tokenizer  vocab.txt 
        self.pos_encoding = positional_encoding(maximum_position_encoding, self.d_model)  #  
        self.dropout = tf.keras.layers.Dropout(dropout_rate)  #  

    def __call__(self, x, training):
        #  x shape = (batch_size, seq_len)
        seq_len = tf.shape(x)[1]
        out = self.embedding(x)  # shape : (batch_size, input_seq_len, d_model)
        out = out * tf.math.sqrt(
            tf.cast(self.d_model, tf.float32))  # x sqrt(d_model)  ,         
        #           
        out = out + self.pos_encoding[:, :seq_len, :]
        out = self.dropout(out, training=training)

        return out  # shape : (batch_size, input_seq_len, d_model)
````
