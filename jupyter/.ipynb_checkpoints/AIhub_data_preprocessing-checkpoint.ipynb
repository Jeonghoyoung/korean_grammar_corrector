```python
import os, json
import librosa
import re
import pandas as pd
import numpy as np
import soundfile as sf
```

```python
path = '../data/aihub_asr_sample'
```

```python
wav = f'{path}/wav/AHAAA21000349.wav'
```

```python
json_data = f'{path}/json/AHAAA21000349.json'
```

```python
wav_data, sr = librosa.load(wav, sr=16000)
```

```python
def json_parser(path):
    category, domain, utt_id, start, end, form, org_form = [], [], [], [], [], [], []
    with open(f'{path}', 'r') as f:
        data = json.load(f)
    for i in range(len(data['utterance'])):
        category.append(data['metadata']['category'])
        domain.append(data['metadata']['domain'])

        utt_id.append(data['utterance'][i]['id'])
        start.append(data['utterance'][i]['start'])
        end.append(data['utterance'][i]['end'])
        form.append(data['utterance'][i]['form'])
        org_form.append(data['utterance'][i]['original_form'])

        df = pd.DataFrame({'category': category,
                             'domain': domain,
                             'id': utt_id,
                             'start': start,
                             'end': end,
                             'form': form,
                             'original_form': org_form})
    return df
```

```python
df = json_parser(json_data)
```

```python
def split_wav(data, sample_rate, start, end):
    start *= sample_rate
    start = int(start)
    end *= sample_rate
    end = int(end)
    return data[start:end]
```

```python
sp_data = split_wav(wav_data, sr, float(df['start'][0]), float(df['end'][0]))
```

```python
sp_data
```

```python
sf.write('../data/test.wav', sp_data, 16000)
```

```python
def save_split_wav(df, wav_data, sample_rate, save_path):
    for i in range(len(df)):
        split_data = split_wav(wav_data, sample_rate, float(df['start'][i]), float(df['end'][i]))
        sf.write(f"{save_path}/{df['id'][i]}.wav", split_data, 16000)
    return None
```

```python
save_split_wav(df,wav_data, sr, '../data/aihub_asr_sample/save_wav')
```

### form 데이터 필터링

    - {laughing}, (()), &name\d, &number\d, &company-name\d 등 의 데이터 제거
        pattern
            1. \{[a-zA-Z]+\}
            2. \@[ㄱ-힣]+\d{0,2}
            3. \&[a-Z]+\d{1,2}\&
            4. \s/\([a-z]+\)$
    
    - 공백으로 치환할 패턴 : \{[a-zA-Z]+\} 
    - 해당 라인 제거 패턴: \@[ㄱ-힣]+\d{0,2} , \&[a-Z]+\d{1,2}\&
    
    
    
    (1) 라인 제거 패턴이 포함된 텍스트는 제거
    (2) 위의 패턴에 의해 제거 되지 않은 텍스트 공백으로 치환 패턴 정제.

```python
test = '안녕하세요 @노래 하는 (()) 선장 @이름2입니다. /(bgm)'
t = "땡 엔쥐 @웃음1"
t2 = "근데 @노래 (()) 들었어"
t3 = '&name1& 정오 뉴스 마칩니다. 시청해 주셔서 고맙습니다.'
```

```python
sub_p = '\{[a-zA-Z]+\}|\@[ㄱ-힣]+|/\([a-z]+\)$'
line_delete_p = '\@[ㄱ-힣]+\d{1,2}|\({2}\){2}|\&[a-z]+\d{1,2}\&|\&[a-z]+\-[a-z]+\d{1,2}\&'
```

```python
re.findall(line_delete_p, t3)
```

```python
def pattern_filter(text):
    if len(re.findall(line_delete_p, text)) > 0:
        return ''
    else:
        return re.sub(sub_p,'',text)
```

```python
pattern_filter(t3)
```

```python
df['filter_form'] = df['form'].apply(lambda x: pattern_filter(x))
df['filter_org_form'] = df['original_form'].apply(lambda x: pattern_filter(x))
```
