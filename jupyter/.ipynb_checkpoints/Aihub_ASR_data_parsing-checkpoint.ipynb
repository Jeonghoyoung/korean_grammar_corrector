### 파싱 대상 데이터
    - metadata
        - category
        - domain
    - utterance
        - start
        - end
        - form
        - original_form

```python
category, domain, utt_id, start, end, form, org_form = [],[],[],[],[],[],[]

df = pd.DataFrame()
for f in flist[:]:
    print(f)
    with open(f'{path}/{f}', 'r') as f:
        data = json.load(f)
    
    for i in range(len(data['utterance'])):
        category.append(data['metadata']['category'])
        domain.append(data['metadata']['domain'])
        
        utt_id.append(data['utterance'][i]['id'])
        start.append(data['utterance'][i]['start'])
        end.append(data['utterance'][i]['end'])
        form.append(data['utterance'][i]['form'])
        org_form.append(data['utterance'][i]['original_form'])
        
        f_df = pd.DataFrame({'category':category,
                            'domain':domain,
                            'id':utt_id,
                            'start':start,
                            'end':end,
                            'form':form,
                            'original_form':org_form})
    print(len(f_df))
    
    df = pd.concat([f_df, df], axis=0)
print(len(df))
```

```python
print(len(category))
print(len(domain))
print(len(utt_id))
print(len(start))
print(len(end))
print(len(form))
print(len(org_form))
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df.to_csv('../data/smaple_json_parsing.csv', index=False, encoding='utf-8-sig')
```
