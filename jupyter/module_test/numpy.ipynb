```python
import numpy as np
```

```python
# ndarray 객체 z의 원소들로부터 정수 부분만 추출
np.random.seed(0)
z = np.random.uniform(0, 10,10)
print(z)
```

```python
z = z.astype(int)
z
```

```python
# 행의 값이 0~4 정수 데이터로 구성된 5X5 2차원 배열 객체 생성
index = np.random.randint(5, size=(5,5))
print(index)
```

```python
# random()함수로 10개의 원소를 생성, 생성된 원소중 최대값을 0으로 변경.
a = np.random.rand(10)
print(a)
a[a == np.max(a)] = int(0)
print(a)
```

```python
# 크기가 10인 랜덤 벡터를 생성하고 벡터의 원소를 오름차순으로 정렬.
b = np.random.rand(10)
b = np.sort(b)
b
```

```python
# z배열 객체에서 빈도수가 가장 높은 값을 찾아 출력
print(z)
val,cnt = np.unique(z, return_counts=True)
idx = np.argmax(cnt)
print(val)
print(cnt)
print(val[idx])
```

```python
# 주사위 100번 던져서 나오는 숫자의 평균을 구하시오.
dice = np.random.randint(1,7, size=100)
print(dice)
np.mean(dice)
```

```python
# 가격이 10000원인 주식이 있고, 이 주식의 일간 수익률은 평균이 0, 표준편차가 1인 정규분포를 따른다고 가정할때, 250일 동안의 주가를 무작위로 생성.
```

```python
diff_x = np.random.normal(0,1,250)
stock_l = [10000]
dif_s = 10000
for x in diff_x:
    dif_s = dif_s + (x*100)
    stock_l.append(dif_s)
print(len(stock_l))
```

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({ '이름': ['영희', '철수', '철수'], '성적': [1, 2, 3]} )
print(df1)
df2 = pd.DataFrame({ '성명': ['영희', '영희', '철수'], '성적2': [4, 5, 6]} )
print(df2)
print(pd.merge(df1, df2, left_on='이름', right_on="성명"))
```
